```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define colors for binary classes
colors = ["#1f77b4", "#ff7f0e"]  # Smooth: blue, Features/Disk: orange

sns.countplot(x='binary_label', data=mapping_df, palette=colors)
plt.xticks([0,1], ["Smooth", "Features/Disk"])
plt.title("Class Distribution (Binary)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define colors for each class
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # blue, orange, green

sns.countplot(x='multiclass_label', data=mapping_df, palette=colors)
plt.xticks([0,1,2], ["Smooth", "Features/Disk", "Artifact"])
plt.title("Class Distribution (Multi-Class)")
plt.show()
```
